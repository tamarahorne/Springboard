```python
import pandas as pd
import sqlite3 as sql
database = 'sqlite_db_pythonsqlite.db'
connection = sql.connect(database)
```

```python
query = "SELECT * FROM Facilities"
```

```python
df = pd.read_sql_query(query,connection)
df
```

```python
queryQ10 = """SELECT name, SUM(revenue) AS total_revenue
FROM(

SELECT name,
((member_slots * membercost) + (guest_slots * guestcost)) AS revenue
FROM (
SELECT name, 
CASE WHEN Bookings.memid != 0 THEN slots ELSE 0 END AS member_slots,
CASE WHEN Bookings.memid = 0 THEN slots ELSE 0 END AS guest_slots,
membercost, guestcost

FROM Bookings
INNER JOIN Facilities on Bookings.facid = Facilities.facid
INNER JOIN Members on Bookings.memid = Members.memid) subquery) outer_subquery
GROUP BY name
HAVING total_revenue < 1000;"""
dfQ10 = pd.read_sql_query(queryQ10,connection)
dfQ10
```

```python
queryQ11 = """
SELECT member_name, recommendedby_name
FROM (
SELECT (m1.surname||', '||m1.firstname) member_name, m1.recommendedby, 
CASE WHEN m2.memid != 0 THEN (m2.surname||', '||m2.firstname) END AS recommendedby_name, 
m2.memid
FROM `Members` as m1
INNER JOIN Members as m2 ON m1.recommendedby = m2.memid
WHERE m1.memid !=0) subquery
ORDER BY member_name;"""
dfQ11 = pd.read_sql_query(queryQ11,connection)
dfQ11
```

```python
#Q12: Find the facilities with their usage by member, but not guests

queryQ12 = """
SELECT name AS facility_name, SUM(slots) AS slots_booked_by_members
FROM (

SELECT b.facid, b.memid, slots, name
FROM Bookings AS b
INNER JOIN Facilities AS f ON b.facid = f.facid
INNER JOIN Members AS m ON b.memid = m.memid
WHERE b.memid != 0) subquery
GROUP BY name;"""
dfQ12 = pd.read_sql_query(queryQ12,connection)
dfQ12
```

```python
#Q13: Find the facilities usage by month, but not guests

queryQ13 = """
SELECT facility_name, month, SUM(slots) AS slots_per_facility
FROM (
    
    SELECT b.facid, b.memid, name AS facility_name, strftime('%m', starttime) as month, slots
    FROM Bookings AS b
    INNER JOIN Facilities AS f ON b.facid = f.facid
    INNER JOIN Members AS m ON b.memid = m.memid WHERE b.memid !=0) subquery
    
GROUP BY facility_name, month;"""
dfQ13 = pd.read_sql_query(queryQ13, connection)
dfQ13
```
